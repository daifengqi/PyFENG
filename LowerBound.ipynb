### Rogers(1995)
At first, for a given asset price $Y$, to obtain a lower bound, we have

$$
E(Y^+)=E(E(Y^+|Z))\ge E(E(Y|Z)^+)
$$

Rogers and Shi think that if we condition the process $X$ on some zero-mean Gaussian variable $Z$, it remains a Gaussian process, and this is the heart of the lower-bound method. Let us write

$$
E(B_t|Z)=m_tZ
$$

We know that

$$
m_t=E(B_tZ)/E(Z^2)
$$

If we take $Z=\int_0^1B_tdt$, then

$$
m_t=3t(2-t)/2
$$

This is an example of conditioning on $Z$.

For an Asian option

$$
\begin{align}
E[Y|Z]&=E[\int_0^1exp(\sigma B_t-\frac{1}{2}\sigma^2t+rt)\mu(dt)|Z] \\
&=\int_0^1exp(\sigma m_tZ-\frac{1}{2}\sigma^2\nu m_t^2+rt)u(dt)
\end{align}
$$

We can also compute the variance to get the upper bound.

### Beisser(1999), Krekel etc.(2004)
When we do research on a basket option, we define

$$
B(T)=\sum_{i=1}^n w_iS_i(T)
$$


Krekel's(2004) paper declaims that they adopt Beisser(1999)'s method that utilizes and idea of Rogers and Shi(1995) introduced for pricing Asian option. By conditioning on the random variable $Z$ and using Jensen's inequality, the price of basket call is estimated in a way,

$$
\begin{align}
E([B(T)-K]^+)&=E(E([B(T)-K]^+|Z)) \\
&\ge E(E([B(T)-K|Z]^+))  \\
&=E([\sum_{i=1}^n w_iE[S_i(T)|Z]-K]^+)
\end{align}
$$

We call the right side as the lower bound of a basket option.

This makes it's possible to condition on single random variable $Z$ to price the option. The authors define that

$$
Z=\frac{\sigma_z}{\sqrt(T)}W(T)=\sum_{i=1}^nw_iS_i(0)\sigma_iW_i(T)
$$

$\sigma_z$ should be appropriately chosen. In contradiction to S_i(T), all conditional expectations $E[S_i(T)|Z]$ are log-normally distributed with resprect to $one$ Brownian motion W(T). Hence, there exists an $x^*$, such that

$$
\sum_{i=1}^n w_iE[S_i(T)|W(T)=x^*]=K
$$

if we find $x^*$, we can define

$$
\widetilde{K}_i=E[S_i(T)|W(T)=x^*]
$$

Finally we get that

$$
\begin{align}
E([\sum_{i=1}^n w_iE[S_i(T)|Z]-K]^+)&=\sum_{i=1}^nw_iE([E[S_i(T)|Z]-\widetilde{K}_i]^+)
\end{align}
$$

### Basket

Let us recall some properties of basket option. 

For $N$ assets, $F_k(T) = F_k(0)exp(-\frac{1}{2}T\sum_{kk}+L_k\sqrt{T}Z)$, where $LL^T=\sum $ and $L_{k*}$ is the k-th row vector of $L$.

Here $L$ is the result of Cholesky decomposition.

Given the volatility and correlation coefficient of the assets, we can obtain $\sum$. Then if we find $V$, where $VV^T=\sum$, then we have

$$
F_k(T) = F_k(0)exp(-\frac{1}{2}T\sum_{kk}+V_k\sqrt{T}Z)
$$

we can conditional on the first random number of $Z$ based on this.

Basically we use PCA of $F$ as $V$ instead of Cholesky's $L$.

```python
import numpy as np
import pandas as pd
# at the current directory
import pyfeng as pf
```

```python
texp = 5
rho = 0.5
o4 = np.ones(4)
sigma = o4 * 0.4
file = 'KrekelEtAl2004-Wilmott-BasketOption.xlsx'
```

# Table 1
## Changing Correlation

```python
df = pd.read_excel(file, sheet_name='1')
df
```

```python
rhos = [0.1, 0.3, 0.5, 0.7, 0.8, 0.95]
result = np.zeros_like(rhos)
for k in range(len(rhos)):
    m = pf.BsmBasketLowerBound(sigma=sigma, cor=rhos[k])
    result[k] = np.round(m.price(100, 100*o4, texp), 2)
np.round(result, 2), np.round(result, 2) - df['Beisser'].values
```

# Table 2
## Changing Strike

```python
df = pd.read_excel(file, sheet_name='2')
df
```

```python
strike = np.arange(50, 151, 10)
m = pf.BsmBasketLowerBound(sigma=sigma, cor=0.5)
result = m.price(strike, spot=100*o4, texp=texp)
np.round(result, 2), np.round(result, 2) - df['Beisser'].values
```

# Table 3
## Changing Forward

```python
df = pd.read_excel(file, sheet_name='3')
df
```

```python
spot = np.arange(50, 151, 10)[:,None]*np.ones(4)
m = pf.BsmBasketLowerBound(sigma=sigma, cor=0.5)
result = [m.price(100, spot=spot[i], texp=texp) for i in range(spot.shape[0])]
np.round(result, 2), np.round(result, 2) - df['Beisser'].values
```
